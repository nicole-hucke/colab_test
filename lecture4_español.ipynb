#**2. Cmo graficar una funcin?** Graficando funciones matemticas

Para visualizar grficamente una funcin matemtica que depende de una variable, podemos crearla utilizando `numpy`.

Por ejemplo, grafiquemos la funcin seno entre 0 y 2 $\pi$.

```python
x = np.linspace(0, 2 * np.pi, 100) # generar 100 puntos entre 0 y 2pi
print(x)
```

Podemos evaluar la funcin seno usando todos los valores en x

```python
y = np.sin(x) # calcular el seno de cada punto
```

Ahora intentemos graficar la funcin seno!

```python
plt.plot(x, y) # x de 0 a 2pi, y = sin(x)
plt.show()
```

En Matplotlib, `plt.show()` se utiliza para mostrar el grfico en la pantalla. Sin llamar a `plt.show()`, es posible que el grfico no se renderice, especialmente en scripts o entornos no interactivos. Asegura que la ventana del grfico se abra y que el grfico se muestre.

El grfico que generamos tiene un par de problemas:

- *Qu estamos graficando exactamente?*
- *Qu representa cada eje?*

Estas son cosas que deben especificarse en cada grfico, ya que pueden no ser claras para el lector. Vamos a aadir el nombre de los ejes y un ttulo a la figura.

```python
plt.plot(x, y)

# agregar etiquetas a los ejes
plt.xlabel('ngulo')
plt.ylabel('Seno')

# agregar ttulo al grfico
plt.title('Funcin Seno')

plt.show()
```

### Mltiples curvas en el mismo grfico

Muchas veces queremos graficar varias funciones o un conjuntos de datos en el mismo grfico. `plt.plot()` puede superponer diferentes curvas simplemente siendo llamado repetidamente en el mismo grfico.

Podemos asignar a cada curva diferentes caractersticas, como `label`, `linestyle` y `color`. De esta manera, es ms fcil diferenciarlas. Para mostrar las etiquetas, necesitamos llamar a `plt.legend()`.

Puedes consultar los diferentes [estilos de lnea](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html) y [colores](https://matplotlib.org/stable/gallery/color/named_colors.html) disponibles en la [documentacin](https://matplotlib.org/stable/) de `matplotlib`.

```python
# primera curva - seno(x)
plt.plot(x, np.sin(x), linestyle="-", label="seno(x)", color="tab:blue")
# segunda curva - seno(2x)
plt.plot(x, np.sin(2 * x), linestyle="--", label="seno(2x)", color="tab:orange")
# tercera curva - seno(3x)
plt.plot(x, np.sin(3 * x), linestyle="-.", label="seno(3x)", color="tab:green")

# agregar una leyenda
plt.legend()

# agregar el nombre del eje x
plt.xlabel('ngulo')
# agregar el nombre del eje y
plt.ylabel('Seno')
# agregar un ttulo
plt.title('Funcin Seno')

plt.show()
```

### Creando Subgrficos

A veces, tener todas estas funciones en un solo grfico puede parecer abarrotado.

Lo que podemos hacer para resolver esto es crear **subgrficos**, donde estamos creando una figura que contiene grficos individuales. Para esto, podemos usar la funcin `subplot` de pyplot.

```python
# Queremos mostrar los tres grficos en una sola fila
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,4)) 
# 1 fila, 3 columnas, y un tamao de 15x4 para que quepan los tres grficos

# primer grfico en ax1
ax1.plot(x, np.sin(x), linestyle="-", label="seno(x)", color="tab:blue")
ax1.set_title('Seno(x)')
ax1.set_xlabel('ngulo')
ax1.set_ylabel('Seno')

# segundo grfico en ax2
ax2.plot(x, np.sin(2 * x), linestyle="--", label="seno(2x)", color="tab:orange")
ax2.set_title('Seno(2x)')
ax2.set_xlabel('ngulo')
ax2.set_ylabel('Seno')

# tercer grfico en ax3
ax3.plot(x, np.sin(3 * x), linestyle="-.", label="seno(3x)", color="tab:green")
ax3.set_title('Seno(3x)')
ax3.set_xlabel('ngulo')
ax3.set_ylabel('Seno')

plt.tight_layout() # ajustar los grficos para que no se superpongan

# mostrar los grficos
plt.show()
```

## **Grfico de Dispersin**

Para demostrar un grfico de dispersin, importemos uno de los conjuntos de datos predeterminados que tiene `seaborn` en su [repositorio de datos](https://github.com/mwaskom/seaborn-data).

Carguemos un conjunto de datos que contiene informacin sobre exoplanetas, como sus perodos orbitales y masas. Para esto, podemos importarlo directamente utilizando `sns.load_dataset()`.

```python
planets = sns.load_dataset('planets') # automaticamente carga el dataset como un dataframe
planets.head() # mostrar las primeras 5 filas del dataframe
```

Vamos a crear un grfico de dispersin entre el perodo orbital y la masa. As es como lo haramos usando `matplotlib`:

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(6, 4))
plt.plot(planets['orbital_period'], planets['mass'], color='teal', marker='o', linestyle='None')

# agregar etiquetas a los ejes y ttulo
plt.xlabel('Periodo Orbital (das)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subndices
plt.title('Periodo Orbital vs Masa de los Planetas')
plt.show()
```

Hay 7 planetas listados en este conjunto de datos, del 1 al 7. Vamos a graficar las caractersticas de cada planeta por color. Para esto, podemos generar un `colormap` y usar un `for loop` para graficar todos los nmeros de los planetas.

Para el colormap, llamaremos a `plt.cm`, que puedes seguir explorando [aqu](https://matplotlib.org/stable/api/cm_api.html). `plt.cm` Nos permite importar colormaps predefinidos. En este ejemplo, usaremos `viridis`, un colormap bastante comn en geociencia. Puedes consultar la [documentacin](https://matplotlib.org/stable/users/explain/colors/colormaps.html) de `matplotlib` para ver otros ejemplos.

```python
plt.figure(figsize=(6, 4))

# gerear un mapa de colores con la misma longitud que los planetas nicos
planet_numbers = planets['number'].unique() # buscar los valores nicos de la columna 'number': 1 a 7
colors = plt.cm.viridis(np.linspace(0, 1, len(planet_numbers))) # # generar un color map con la misma longitud que los planetas nicos

# graficar cada planeta con un color diferente
for i, planet in enumerate(planet_numbers):
    planet_data = planets[planets['number'] == planet] # datos de cada planeta individualmente
    # planet_data se establece asegurndose de que la columna number sea igual al nmero del planeta
    plt.plot(planet_data['orbital_period'], planet_data['mass'], color=colors[i],
            marker='o', linestyle='None', label= planet)

# agregar etiquetas a los ejes y ttulo
plt.xlabel('Periodo Orbital (das)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subndices
plt.title('Periodo Orbital vs Masa de los Planetas')
plt.legend()
plt.show()
```

Como pudiste ver, hay muchos pasos para poder hacer esto en `matplotlib`, y puede volverse complicado y engorroso rpidamente. Afortunadamente, `seaborn` puede crear estos colormaps automticamente utilizando el argumento `hue` en la funcin `sns.scatterplot()`.

En lugar de un colormap, `seaborn` trabaja con `paletas`, que puedes especificar como un argumento al graficar los datos. Para ms informacin sobre las `paletas de colores`, puedes consultar la [documentacin aqu](https://seaborn.pydata.org/tutorial/color_palettes.html). Tambin puedes asignar un marcador diferente aadiendo el argumento `style`.

```python
plt.figure(figsize=(6, 4))
sns.set_style("whitegrid")
sns.scatterplot(data=planets, x="orbital_period", y="mass", hue="number", palette='deep', style="number")

plt.xlabel('Periodo Orbital (das)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subndices
plt.title('Periodo Orbital vs Masa de los Planetas')
plt.legend()
plt.show()
```

Tambin podemos categorizar cada nmero de planeta por el tamao del marcador. Aqu tienes un ejemplo donde usamos `size` para diferenciar entre los nmeros de los planetas.

Adems, si queremos cambiar los lmites en el eje X o Y, podemos hacerlo con `plt.xlim` y `plt.ylim`, respectivamente.

```python
plt.figure(figsize=(6, 4))
sns.set_style("darkgrid")
sns.scatterplot(data=planets, x="orbital_period", y="mass", hue="number", palette='crest', size="number")

plt.xlabel('Periodo Orbital (das)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subndices
plt.title('Periodo Orbital vs Masa de los Planetas')

plt.xlim(-200, 5000)  # determinar los lmites del eje x
plt.ylim(-1, 24)  # determinar los lmites del eje y

plt.legend()
plt.savefig('plots/periodo_orbital_vs_masa_planetas.png', dpi=300)
plt.show()
```

## **Grfico Circular**

Tambin conocido como **grfico pastel**. La funcin para graficar estos en `matplotlib` es `plt.pie()`. La documentacin y otros ejemplos sobre esta funcin se pueden encontrar [aqu](https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html#sphx-glr-gallery-pie-and-polar-charts-pie-features-py). No hay una funcin de grfico de pastel en `seaborn`.

Vamos a crear un `grfico circular` que represente los diferentes mtodos utilizados para obtener los datos de los planetas. Para esto, necesitamos saber cuntas mediciones existen por cada mtodo y agruparemos todos los mtodos que tengan menos de 10 conteos.

```python
# contar cuntas mediciones hay para cada mtodo
method_counts = planets['method'].value_counts()
method_counts
```

```python
# identificar los mtodos con menos de 10 conteos
low_count_methods = method_counts[method_counts < 10]
other_count = low_count_methods.sum()

# filtrar los mtodos con menos de 10 conteos (otros)
filtered_method_counts = method_counts[method_counts >= 10]

# agregar la categora "Otros"
filtered_method_counts['Other'] = other_count
filtered_method_counts
```

Ahora podemos continuar y graficar!

El argumento `autopct` aadir los porcentajes de las respectivas porciones, `startangle` es el ngulo donde comenzar la primera porcin, y `color` se utiliza para elegir los colores individuales o el colormap que queremos para el grfico.

```python
data = filtered_method_counts.values # obtener los valores de la serie method_counts
labels = filtered_method_counts.index # el ndice corresponde a los nombres de los mtodos

sns.reset_orig() # resetear el estilo de los grficos
plt.figure(figsize=(10, 5))
plt.pie(data, labels=labels, autopct='%.0f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Measurement Methods')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/grfico_circular_metodos.png', dpi=300)
plt.show()
```

## **Histograma**

Los histogramas son grficos que representan una distribucin de valores numricos como una serie de barras, los cuales se agrupan en intervalos de clase, tambin conocidos como `bins`. La altura de una barra indica la frecuencia de valores dentro de un bin determinado. Tanto `matplotlib` como `seaborn` pueden crear histogramas.

La funcin de `matplotlib` para esto es `plt.hist()`, donde necesitamos especificar el arreglo de datos que se va a graficar. Tambin podemos especificar cuntos `bins` existirn y si es que queremos una `funcin de densidad` (para que el rea bajo el histograma se integre a 1). Ms informacin sobre la funcin `plt.hist()` se puede encontrar en la [documentacin](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html).

Vamos a graficar un histograma de la duracin de las explosiones de giseres!

```python
geyser = sns.load_dataset('geyser')
geyser.head() # mostrar las primeras 5 filas del dataframe
```

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(6, 4))

plt.hist(geyser['duration'], bins=20, density=False, color='skyblue') # graficar el histograma

plt.xlabel('Duracin (minutos)') 
plt.ylabel('Frecuencia')
plt.title('Duraciones de las erupciones de geysers')
plt.show()
```

Para `seaborn`, la funcin es `sns.histplot()`, donde adems puedes especificar si deseas que las barras estn en el eje `X` o `Y`, agregar una estimacin de densidad de kernel (kde), o dibujar mltiples histogramas con el mapeo `hue`. Para ver otros argumentos, puedes consultar la documentacin [aqu](https://seaborn.pydata.org/generated/seaborn.histplot.html).

```python
plt.figure(figsize=(6, 4))
sns.set_style("darkgrid")
sns.histplot(data=geyser, x='duration', bins=20, kde=False)

plt.xlabel('Duracin (minutos)') 
plt.ylabel('Frecuencia')
plt.title('Duraciones de las erupciones de geysers')
plt.show()
```

Grafiquemos las distribuciones de las erupciones *largas* y *cortas*, as como la KDE para ambas:

```python
plt.figure(figsize=(8, 4))
sns.set_style("darkgrid")
sns.histplot(data=geyser, x='duration', bins=20, kde=True, hue='kind')

plt.xlabel('Duracin (minutos)') 
plt.ylabel('Frecuencia')
plt.title('Duraciones de las erupciones de geysers')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/erupcion_geyser_histograma.png', dpi=300)
plt.show()
```

## **Grfico de Barras**

Este tipo de grficos puede parecer similar a los histogramas, pero la diferencia entre ellos es que los **grficos de barras comparan variables discretas o categricas**, mientras que el histograma muestra la distribucin de frecuencia de las variables en un conjunto de datos.

Para `matplotlib`, la funcin es `plt.bar`, y puedes encontrar ms informacin [aqu](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html).

Hagamos un `grfico de barras` que muestre la duracin promedio de erupcin de cada *tipo* de giser.

```python
# primero necesitamos calcular la duracin media de las erupciones largas y cortas
long_average = geyser[geyser['kind'] == 'long']['duration'].mean() # duracin media de las erupciones largas
short_average = geyser[geyser['kind'] == 'short']['duration'].mean() # duracin media de las erupciones cortas
print(short_average, long_average)

# agregar los valores a un dataframe
average_duration = pd.DataFrame({'kind': ['long', 'short'], 'duration': [long_average, short_average]})
average_duration
```

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(6, 4))
plt.bar(average_duration['kind'], average_duration['duration'])
plt.xlabel('Tipo de erupcin')
plt.ylabel('Duracin promedio (minutos)')
plt.title('Duracin promedio de las erupciones de geysers')
plt.show()
```

Para obtener los tiempos de duracin promedio, tuvimos que calcularlos nosotros mismos y **luego** agregarlos al grfico. Un beneficio de `seaborn` es que agrupa los datos por la columna *kind* y calcula la `media` (u otro estimador si se especifica) de la columna *duration* para cada grupo.

As que, para graficar la duracin promedio de los tipos de erupcin de giseres, podemos hacerlo directamente usando la funcin `sns.barplot`, sobre la cual puedes encontrar ms informacin [aqu](https://seaborn.pydata.org/generated/seaborn.barplot.html).

```python
sns.set_style("darkgrid")
sns.barplot(data=geyser, x="kind", y="duration")
plt.xlabel('Tipo de erupcin')
plt.ylabel('Duracin promedio (minutos)')
plt.title('Duracin promedio de las erupciones de geysers')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/erupcion_geyser_barras.png', dpi=300)
plt.show()
```

## **Diagrama de caja**

Un diagrama de caja, tambin conocido como `box-and-whisker plot`, es un grfico que muestra la distribucin de datos numricos usando una caja y lneas. Es una herramienta til para resumir datos y comparar diferentes conjuntos de datos. Si no has visto uno antes, as es como se interpretan:

box_plot.JPG


Volvamos a nuestros datos de temperatura media del aire de [Berkeley Earth](https://berkeleyearth.org/) y creemos un `diagrama de caja` de la distribucin de todas las temperaturas desde 1850 hasta 2020.

Para `matplotlib`, usaramos la funcin `plt.boxplot`, sobre la cual puedes encontrar ms informacin [aqu](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html).

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(7, 5))
plt.boxplot(temperatura)
plt.ylabel('Temperatura (C)')
plt.title('Temperatura media anual de la Tierra de 1850 a 2020')
plt.show()
```

Ms all de los bigotes (`whiskers`), hay algunos puntos de datos que aparecen como **puntos**. Estos puntos se consideran `valores atpicos`. En un diagrama de caja, los `valores atpicos` son puntos de datos que caen fuera del rango definido por los bigotes. Para `matplotlib`, los bigotes se extienden hasta 1.5 veces el rango intercuartil (IQR) desde el primer y tercer cuartil. Cualquier punto de datos que est fuera de este rango se considera un `valor atpico` y se grafica como puntos individuales.

Sin embargo, dado que estamos graficando un rango tan amplio de aos y temperaturas, ponerlos todos en un solo diagrama de caja no es representativo, especialmente si queremos ver alguna tendencia a lo largo del tiempo.

Vamos a crear diagramas de caja para las temperaturas **cada 10 aos** (agrupando de 1850 a 1859, de 1860 a 1869... y de 2010 a 2019), donde cada dcada tiene su propio diagrama de caja.

```python
# convirtiendo los datos de temperatura a un dataframe
temperature_data = pd.DataFrame(temperature_data, columns=['Year', 'Temperature'])
# convertir el ao a un entero
temperature_data['Year'] = temperature_data['Year'].astype(int) # no hay decimales en los aos
# crear una nueva columna para la dcada
temperature_data['Decade'] = (temperature_data['Year'] // 10) * 10 # redondear a la dcada ms cercana
temperature_data.head(11)
```

`matplotlib` tiene dificultades para crear diagramas de caja gestionando el agrupamiento y la representacin de los datos de manera interna, por lo que necesitamos agruparlos nosotros mismos:

```python
# agrupar las temperaturas por dcada
grouped_temperatures = temperature_data.groupby('Decade')['Temperature'].apply(list)
grouped_temperatures
```

Ahora podemos usar la funcin `plt.boxplot` de `matplotlib` para crear nuestro grfico:

```python
plt.figure(figsize=(10, 5))
plt.boxplot(grouped_temperatures, labels=grouped_temperatures.index) # el indice es la dcada
plt.title('Distribucin de la temperatura media anual por dcada')
plt.xlabel('Dcada')
plt.ylabel('Temperatura (C)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Este grfico es mucho ms representativo de la variacin de la temperatura a lo largo del tiempo, as como de cmo cambia la distribucin de cada dcada!

Si quisiramos evitar agrupar manualmente las temperaturas, `seaborn` es capaz de agruparlas internamente y crear diagramas de caja utilizando la funcin `sns.boxplot`. Puedes encontrar ms informacin sobre ello [aqu](https://seaborn.pydata.org/generated/seaborn.boxplot.html). Vamos a crear el mismo grfico que hicimos anteriormente:

```python
plt.figure(figsize=(12, 5))
sns.set_style("darkgrid")
sns.boxplot(data=temperature_data, x='Decade', y='Temperature')
plt.title('Distribucin de la temperatura media anual por dcada')
plt.xlabel('Dcada')
plt.ylabel('Temperatura (C)')
plt.xticks(rotation=45)
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/temperaturas_caja.png', dpi=300)
plt.show()
```

## **Grfico de Violn**

Similar al `diagrama de caja`, el `grfico de violn` muestra la distribucin de datos numricos utilizando **curvas de densidad**. Es una herramienta til para visualizar y comparar distribuciones entre grupos, e identificar mnimos, mximos y colas en los datos.

Vamos a crear `grficos de violn` de los datos de temperatura que graficamos anteriormente.

En `matplotlib`, utilizamos `plt.violinplot`, donde puedes especificar si queremos mostrar la `media` o `mediana` de cada distribucin. Los parmetros de entrada son diferentes de los de `plt.boxplot`, y no podemos graficar las etiquetas directamente en la funcin. Puedes encontrar ms informacin sobre esta funcin y sus argumentos [aqu](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html).

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(12, 5))
plt.violinplot(grouped_temperatures, showmedians=True, showmeans=False) # mostrar la mediana pero no la media
plt.title('Distribucin de la temperatura media anual por dcada')
plt.xlabel('Dcada')
plt.ylabel('Temperatura (C)')
plt.xticks(ticks=range(1, len(grouped_temperatures.index) + 1), labels=grouped_temperatures.index, # etiquetas de las dcadas
            rotation=45, ha='right')
plt.show()
```

En `seaborn`, la funcin que utilizamos es `sns.violinplot`, que puede tener considerablemente ms argumentos que en `matplotlib`. Puedes encontrar toda la informacin sobre esta funcin [aqu](https://seaborn.pydata.org/generated/seaborn.violinplot.html).

Una ventaja de esta funcin es que, al igual que los diagramas de caja, es capaz de agruparlos internamente, por lo que podemos darle directamente el `temperature_data` original en lugar de los `grouped_temperatures`.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
sns.violinplot(data=temperature_data, x='Decade', y='Temperature')
plt.title('Distribucin de la temperatura media anual por dcada')
plt.xlabel('Dcada')
plt.ylabel('Temperatura (C)')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/temperatura_violin.png', dpi=300)
plt.show()
```

### **Cundo Usar un Diagrama de Violn en Lugar de un Diagrama de Caja?**

Un `diagrama de violn` es til cuando queremos mostrar la **forma de la distribucin** junto con los detalles que ofrece un diagrama de caja. Es particularmente valioso para conjuntos de datos ms grandes, ya que proporciona ms informacin sobre la distribucin de los datos.

Por otro lado, un `diagrama de caja` es beneficioso para resaltar la **tendencia central** y la **dispersin de los datos**. Tambin es til para identificar cualquier `valor atpico` presente en el conjunto de datos.


## **Mapas de Calor**

Estos son una representacin **bidimensional** de datos en la que varios valores se representan mediante colores. Usar `mapas de calor` es una buena manera de representar datos de una manera ms visual. Todos los `mapas de calor` comparten una cosa en comn: utilizan **diferentes colores para representar diferentes valores** y para comunicar las relaciones que pueden existir entre las variables.

En `matplotlib`, la funcin que puede hacer esto es `plt.imshow`, que no se usa exclusivamente para crear mapas de calor. `plt.imshow` puede mostrar imgenes importadas, interpolar imgenes, crear grficos de contorno, etc. Puedes encontrar ms informacin sobre esta funcin [aqu](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html), as como ejemplos adicionales sobre cmo [graficar imgenes](https://matplotlib.org/stable/gallery/images_contours_and_fields/image_demo.html#sphx-glr-gallery-images-contours-and-fields-image-demo-py).

As es como usaramos `plt.imshow`:

```python
data = np.random.random((12, 12)) # generar datos aleatorios de 12x12

plt.figure(figsize=(6, 6))
plt.imshow(data, cmap='viridis', interpolation='nearest') # grafiar los datos como una imagen e interpolar los valores
plt.grid(False)  # desactivar la cuadrcula
plt.colorbar() # mostrar la barra de colores
plt.title("Mapa de calor aleatorio")
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.show()
```

Dado que `plt.imshow` es una funcin tan verstil, el cdigo para crear un mapa de calor especfico anotado puede volverse bastante largo. Aqu hay algunos [ejemplos](https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html#sphx-glr-gallery-images-contours-and-fields-image-annotated-heatmap-py) sobre cmo hacerlo.

Afortunadamente, `seaborn` tiene la funcin `sns.heatmap`, que puede graficar datos rectangulares como una matriz codificada por colores directamente. Como ejemplo, importemos un dataframe del [repositorio de datos de seaborn](https://github.com/mwaskom/seaborn-data/tree/master) con datos sobre las caractersticas de los diamantes y creemos un mapa de calor que nos informe sobre el `precio` de un diamante segn su `corte`, `color` y `claridad`.

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

Primero separaremos por `corte` y `claridad`. Para esto, utilizaremos el mtodo `pivot_table` de `pandas`, donde las filas estarn indexadas por `corte`, las columnas estarn definidas por `claridad` y los valores de color del mapa de calor sern el `precio` del diamante. Puedes encontrar ms informacin sobre el `pivot_table` en la documentacin de `pandas` [aqu](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html).


```python
diamond_cut_clarity = diamonds.pivot_table(index="cut", columns="clarity", values="price")
diamond_cut_clarity
```

Ahora podemos avanzar y trazar nuestro mapa de calor usando `sns.heatmap`!

Podemos agregar el valor del precio a la matriz aadiendo el argumento `annot=True`. Adems, podemos elegir el `colormap` de nuestro mapa de calor con `cmap`. Puedes encontrar ms informacin sobre esta funcin [aqu](https://seaborn.pydata.org/generated/seaborn.heatmap.html).

```python
plt.figure(figsize=(12, 6))
sns.heatmap(diamond_cut_clarity, annot=True, cmap='cubehelix', fmt='.0f')
# fmt='.0f' es para formatear la anotacin a 0 decimales
plt.title('Precio promedio de diamantes por claridad y corte')
plt.xlabel('Claridad')
plt.ylabel('Corte')
plt.show()
```

Vamos a crear un subplot donde podemos crear mapas de calor que muestren el `precio` de un diamante en funcin del `corte`, `color` y `claridad`:

```python
diamond_cut_color = diamonds.pivot_table(index="cut", columns="color", values="price")
diamond_clarity_color = diamonds.pivot_table(index="clarity", columns="color", values="price")

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5)) # 1 fila, 3 columnas

# corte y claridad
sns.heatmap(diamond_cut_clarity, annot=True, fmt='.0f', ax=ax1)
ax1.set_title('Precio de los diamantes por corte y claridad')
ax1.set_xlabel('Claridad')
ax1.set_ylabel('Corte')

# cut and color
sns.heatmap(diamond_cut_color, annot=True, fmt='.0f', ax=ax2)
ax2.set_title('Precio de los diamantes por corte y color')
ax2.set_xlabel('Color')
ax2.set_ylabel('Corte')

# clarity and color
sns.heatmap(diamond_clarity_color, annot=True, fmt='.0f', ax=ax3)
ax3.set_title('Precio de los diamantes por claridad y color')
ax3.set_xlabel('Color')
ax3.set_ylabel('Claridad')

plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/diamantes_mapacalor.png', dpi=300)
plt.show()
```
