#**2. ¿Cómo graficar una función?** Graficando funciones matemáticas

Para visualizar gráficamente una función matemática que depende de una variable, podemos crearla utilizando `numpy`.

Por ejemplo, grafiquemos la función seno entre 0 y 2 $\pi$.

```python
x = np.linspace(0, 2 * np.pi, 100) # generar 100 puntos entre 0 y 2pi
print(x)
```

Podemos evaluar la función seno usando todos los valores en x

```python
y = np.sin(x) # calcular el seno de cada punto
```

¡Ahora intentemos graficar la función seno!

```python
plt.plot(x, y) # x de 0 a 2pi, y = sin(x)
plt.show()
```

En Matplotlib, `plt.show()` se utiliza para mostrar el gráfico en la pantalla. Sin llamar a `plt.show()`, es posible que el gráfico no se renderice, especialmente en scripts o entornos no interactivos. Asegura que la ventana del gráfico se abra y que el gráfico se muestre.

El gráfico que generamos tiene un par de problemas:

- *¿Qué estamos graficando exactamente?*
- *¿Qué representa cada eje?*

Estas son cosas que deben especificarse en cada gráfico, ya que pueden no ser claras para el lector. Vamos a añadir el nombre de los ejes y un título a la figura.

```python
plt.plot(x, y)

# agregar etiquetas a los ejes
plt.xlabel('Ángulo')
plt.ylabel('Seno')

# agregar título al gráfico
plt.title('Función Seno')

plt.show()
```

### Múltiples curvas en el mismo gráfico

Muchas veces queremos graficar varias funciones o un conjuntos de datos en el mismo gráfico. `plt.plot()` puede superponer diferentes curvas simplemente siendo llamado repetidamente en el mismo gráfico.

Podemos asignar a cada curva diferentes características, como `label`, `linestyle` y `color`. De esta manera, es más fácil diferenciarlas. Para mostrar las etiquetas, necesitamos llamar a `plt.legend()`.

Puedes consultar los diferentes [estilos de línea](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html) y [colores](https://matplotlib.org/stable/gallery/color/named_colors.html) disponibles en la [documentación](https://matplotlib.org/stable/) de `matplotlib`.

```python
# primera curva - seno(x)
plt.plot(x, np.sin(x), linestyle="-", label="seno(x)", color="tab:blue")
# segunda curva - seno(2x)
plt.plot(x, np.sin(2 * x), linestyle="--", label="seno(2x)", color="tab:orange")
# tercera curva - seno(3x)
plt.plot(x, np.sin(3 * x), linestyle="-.", label="seno(3x)", color="tab:green")

# agregar una leyenda
plt.legend()

# agregar el nombre del eje x
plt.xlabel('Ángulo')
# agregar el nombre del eje y
plt.ylabel('Seno')
# agregar un título
plt.title('Función Seno')

plt.show()
```

### Creando Subgráficos

A veces, tener todas estas funciones en un solo gráfico puede parecer abarrotado.

Lo que podemos hacer para resolver esto es crear **subgráficos**, donde estamos creando una figura que contiene gráficos individuales. Para esto, podemos usar la función `subplot` de pyplot.

```python
# Queremos mostrar los tres gráficos en una sola fila
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,4)) 
# 1 fila, 3 columnas, y un tamaño de 15x4 para que quepan los tres gráficos

# primer gráfico en ax1
ax1.plot(x, np.sin(x), linestyle="-", label="seno(x)", color="tab:blue")
ax1.set_title('Seno(x)')
ax1.set_xlabel('Ángulo')
ax1.set_ylabel('Seno')

# segundo gráfico en ax2
ax2.plot(x, np.sin(2 * x), linestyle="--", label="seno(2x)", color="tab:orange")
ax2.set_title('Seno(2x)')
ax2.set_xlabel('Ángulo')
ax2.set_ylabel('Seno')

# tercer gráfico en ax3
ax3.plot(x, np.sin(3 * x), linestyle="-.", label="seno(3x)", color="tab:green")
ax3.set_title('Seno(3x)')
ax3.set_xlabel('Ángulo')
ax3.set_ylabel('Seno')

plt.tight_layout() # ajustar los gráficos para que no se superpongan

# mostrar los gráficos
plt.show()
```

## **Gráfico de Dispersión**

Para demostrar un gráfico de dispersión, importemos uno de los conjuntos de datos predeterminados que tiene `seaborn` en su [repositorio de datos](https://github.com/mwaskom/seaborn-data).

Carguemos un conjunto de datos que contiene información sobre exoplanetas, como sus períodos orbitales y masas. Para esto, podemos importarlo directamente utilizando `sns.load_dataset()`.

```python
planets = sns.load_dataset('planets') # automaticamente carga el dataset como un dataframe
planets.head() # mostrar las primeras 5 filas del dataframe
```

Vamos a crear un gráfico de dispersión entre el período orbital y la masa. Así es como lo haríamos usando `matplotlib`:

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(6, 4))
plt.plot(planets['orbital_period'], planets['mass'], color='teal', marker='o', linestyle='None')

# agregar etiquetas a los ejes y título
plt.xlabel('Periodo Orbital (días)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subíndices
plt.title('Periodo Orbital vs Masa de los Planetas')
plt.show()
```

Hay 7 planetas listados en este conjunto de datos, del 1 al 7. Vamos a graficar las características de cada planeta por color. Para esto, podemos generar un `colormap` y usar un `for loop` para graficar todos los números de los planetas.

Para el colormap, llamaremos a `plt.cm`, que puedes seguir explorando [aquí](https://matplotlib.org/stable/api/cm_api.html). `plt.cm` Nos permite importar colormaps predefinidos. En este ejemplo, usaremos `viridis`, un colormap bastante común en geociencia. Puedes consultar la [documentación](https://matplotlib.org/stable/users/explain/colors/colormaps.html) de `matplotlib` para ver otros ejemplos.

```python
plt.figure(figsize=(6, 4))

# gerear un mapa de colores con la misma longitud que los planetas únicos
planet_numbers = planets['number'].unique() # buscar los valores únicos de la columna 'number': 1 a 7
colors = plt.cm.viridis(np.linspace(0, 1, len(planet_numbers))) # # generar un color map con la misma longitud que los planetas únicos

# graficar cada planeta con un color diferente
for i, planet in enumerate(planet_numbers):
    planet_data = planets[planets['number'] == planet] # datos de cada planeta individualmente
    # planet_data se establece asegurándose de que la columna number sea igual al número del planeta
    plt.plot(planet_data['orbital_period'], planet_data['mass'], color=colors[i],
            marker='o', linestyle='None', label= planet)

# agregar etiquetas a los ejes y título
plt.xlabel('Periodo Orbital (días)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subíndices
plt.title('Periodo Orbital vs Masa de los Planetas')
plt.legend()
plt.show()
```

Como pudiste ver, hay muchos pasos para poder hacer esto en `matplotlib`, y puede volverse complicado y engorroso rápidamente. Afortunadamente, `seaborn` puede crear estos colormaps automáticamente utilizando el argumento `hue` en la función `sns.scatterplot()`.

En lugar de un colormap, `seaborn` trabaja con `paletas`, que puedes especificar como un argumento al graficar los datos. Para más información sobre las `paletas de colores`, puedes consultar la [documentación aquí](https://seaborn.pydata.org/tutorial/color_palettes.html). También puedes asignar un marcador diferente añadiendo el argumento `style`.

```python
plt.figure(figsize=(6, 4))
sns.set_style("whitegrid")
sns.scatterplot(data=planets, x="orbital_period", y="mass", hue="number", palette='deep', style="number")

plt.xlabel('Periodo Orbital (días)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subíndices
plt.title('Periodo Orbital vs Masa de los Planetas')
plt.legend()
plt.show()
```

También podemos categorizar cada número de planeta por el tamaño del marcador. Aquí tienes un ejemplo donde usamos `size` para diferenciar entre los números de los planetas.

Además, si queremos cambiar los límites en el eje X o Y, podemos hacerlo con `plt.xlim` y `plt.ylim`, respectivamente.

```python
plt.figure(figsize=(6, 4))
sns.set_style("darkgrid")
sns.scatterplot(data=planets, x="orbital_period", y="mass", hue="number", palette='crest', size="number")

plt.xlabel('Periodo Orbital (días)')	
plt.ylabel('Masa $10^{24}$*(kg)') # usar LaTeX para escribir subíndices
plt.title('Periodo Orbital vs Masa de los Planetas')

plt.xlim(-200, 5000)  # determinar los límites del eje x
plt.ylim(-1, 24)  # determinar los límites del eje y

plt.legend()
plt.savefig('plots/periodo_orbital_vs_masa_planetas.png', dpi=300)
plt.show()
```

## **Gráfico Circular**

También conocido como **gráfico pastel**. La función para graficar estos en `matplotlib` es `plt.pie()`. La documentación y otros ejemplos sobre esta función se pueden encontrar [aquí](https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html#sphx-glr-gallery-pie-and-polar-charts-pie-features-py). No hay una función de gráfico de pastel en `seaborn`.

Vamos a crear un `gráfico circular` que represente los diferentes métodos utilizados para obtener los datos de los planetas. Para esto, necesitamos saber cuántas mediciones existen por cada método y agruparemos todos los métodos que tengan menos de 10 conteos.

```python
# contar cuántas mediciones hay para cada método
method_counts = planets['method'].value_counts()
method_counts
```

```python
# identificar los métodos con menos de 10 conteos
low_count_methods = method_counts[method_counts < 10]
other_count = low_count_methods.sum()

# filtrar los métodos con menos de 10 conteos (otros)
filtered_method_counts = method_counts[method_counts >= 10]

# agregar la categoría "Otros"
filtered_method_counts['Other'] = other_count
filtered_method_counts
```

¡Ahora podemos continuar y graficar!

El argumento `autopct` añadirá los porcentajes de las respectivas porciones, `startangle` es el ángulo donde comenzará la primera porción, y `color` se utiliza para elegir los colores individuales o el colormap que queremos para el gráfico.

```python
data = filtered_method_counts.values # obtener los valores de la serie method_counts
labels = filtered_method_counts.index # el índice corresponde a los nombres de los métodos

sns.reset_orig() # resetear el estilo de los gráficos
plt.figure(figsize=(10, 5))
plt.pie(data, labels=labels, autopct='%.0f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Measurement Methods')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/gráfico_circular_metodos.png', dpi=300)
plt.show()
```

## **Histograma**

Los histogramas son gráficos que representan una distribución de valores numéricos como una serie de barras, los cuales se agrupan en intervalos de clase, también conocidos como `bins`. La altura de una barra indica la frecuencia de valores dentro de un bin determinado. Tanto `matplotlib` como `seaborn` pueden crear histogramas.

La función de `matplotlib` para esto es `plt.hist()`, donde necesitamos especificar el arreglo de datos que se va a graficar. También podemos especificar cuántos `bins` existirán y si es que queremos una `función de densidad` (para que el área bajo el histograma se integre a 1). Más información sobre la función `plt.hist()` se puede encontrar en la [documentación](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html).

¡Vamos a graficar un histograma de la duración de las explosiones de géiseres!

```python
geyser = sns.load_dataset('geyser')
geyser.head() # mostrar las primeras 5 filas del dataframe
```

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(6, 4))

plt.hist(geyser['duration'], bins=20, density=False, color='skyblue') # graficar el histograma

plt.xlabel('Duración (minutos)') 
plt.ylabel('Frecuencia')
plt.title('Duraciones de las erupciones de geysers')
plt.show()
```

Para `seaborn`, la función es `sns.histplot()`, donde además puedes especificar si deseas que las barras estén en el eje `X` o `Y`, agregar una estimación de densidad de kernel (kde), o dibujar múltiples histogramas con el mapeo `hue`. Para ver otros argumentos, puedes consultar la documentación [aquí](https://seaborn.pydata.org/generated/seaborn.histplot.html).

```python
plt.figure(figsize=(6, 4))
sns.set_style("darkgrid")
sns.histplot(data=geyser, x='duration', bins=20, kde=False)

plt.xlabel('Duración (minutos)') 
plt.ylabel('Frecuencia')
plt.title('Duraciones de las erupciones de geysers')
plt.show()
```

Grafiquemos las distribuciones de las erupciones *largas* y *cortas*, así como la KDE para ambas:

```python
plt.figure(figsize=(8, 4))
sns.set_style("darkgrid")
sns.histplot(data=geyser, x='duration', bins=20, kde=True, hue='kind')

plt.xlabel('Duración (minutos)') 
plt.ylabel('Frecuencia')
plt.title('Duraciones de las erupciones de geysers')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/erupcion_geyser_histograma.png', dpi=300)
plt.show()
```

## **Gráfico de Barras**

Este tipo de gráficos puede parecer similar a los histogramas, pero la diferencia entre ellos es que los **gráficos de barras comparan variables discretas o categóricas**, mientras que el histograma muestra la distribución de frecuencia de las variables en un conjunto de datos.

Para `matplotlib`, la función es `plt.bar`, y puedes encontrar más información [aquí](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html).

Hagamos un `gráfico de barras` que muestre la duración promedio de erupción de cada *tipo* de géiser.

```python
# primero necesitamos calcular la duración media de las erupciones largas y cortas
long_average = geyser[geyser['kind'] == 'long']['duration'].mean() # duración media de las erupciones largas
short_average = geyser[geyser['kind'] == 'short']['duration'].mean() # duración media de las erupciones cortas
print(short_average, long_average)

# agregar los valores a un dataframe
average_duration = pd.DataFrame({'kind': ['long', 'short'], 'duration': [long_average, short_average]})
average_duration
```

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(6, 4))
plt.bar(average_duration['kind'], average_duration['duration'])
plt.xlabel('Tipo de erupción')
plt.ylabel('Duración promedio (minutos)')
plt.title('Duración promedio de las erupciones de geysers')
plt.show()
```

Para obtener los tiempos de duración promedio, tuvimos que calcularlos nosotros mismos y **luego** agregarlos al gráfico. Un beneficio de `seaborn` es que agrupa los datos por la columna *kind* y calcula la `media` (u otro estimador si se especifica) de la columna *duration* para cada grupo.

Así que, para graficar la duración promedio de los tipos de erupción de géiseres, podemos hacerlo directamente usando la función `sns.barplot`, sobre la cual puedes encontrar más información [aquí](https://seaborn.pydata.org/generated/seaborn.barplot.html).

```python
sns.set_style("darkgrid")
sns.barplot(data=geyser, x="kind", y="duration")
plt.xlabel('Tipo de erupción')
plt.ylabel('Duración promedio (minutos)')
plt.title('Duración promedio de las erupciones de geysers')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/erupcion_geyser_barras.png', dpi=300)
plt.show()
```

## **Diagrama de caja**

Un diagrama de caja, también conocido como `box-and-whisker plot`, es un gráfico que muestra la distribución de datos numéricos usando una caja y líneas. Es una herramienta útil para resumir datos y comparar diferentes conjuntos de datos. Si no has visto uno antes, así es como se interpretan:

box_plot.JPG


Volvamos a nuestros datos de temperatura media del aire de [Berkeley Earth](https://berkeleyearth.org/) y creemos un `diagrama de caja` de la distribución de todas las temperaturas desde 1850 hasta 2020.

Para `matplotlib`, usaríamos la función `plt.boxplot`, sobre la cual puedes encontrar más información [aquí](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html).

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(7, 5))
plt.boxplot(temperatura)
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura media anual de la Tierra de 1850 a 2020')
plt.show()
```

Más allá de los bigotes (`whiskers`), hay algunos puntos de datos que aparecen como **puntos**. Estos puntos se consideran `valores atípicos`. En un diagrama de caja, los `valores atípicos` son puntos de datos que caen fuera del rango definido por los bigotes. Para `matplotlib`, los bigotes se extienden hasta 1.5 veces el rango intercuartil (IQR) desde el primer y tercer cuartil. Cualquier punto de datos que esté fuera de este rango se considera un `valor atípico` y se grafica como puntos individuales.

Sin embargo, dado que estamos graficando un rango tan amplio de años y temperaturas, ponerlos todos en un solo diagrama de caja no es representativo, especialmente si queremos ver alguna tendencia a lo largo del tiempo.

Vamos a crear diagramas de caja para las temperaturas **cada 10 años** (agrupando de 1850 a 1859, de 1860 a 1869... y de 2010 a 2019), donde cada década tiene su propio diagrama de caja.

```python
# convirtiendo los datos de temperatura a un dataframe
temperature_data = pd.DataFrame(temperature_data, columns=['Year', 'Temperature'])
# convertir el año a un entero
temperature_data['Year'] = temperature_data['Year'].astype(int) # no hay decimales en los años
# crear una nueva columna para la década
temperature_data['Decade'] = (temperature_data['Year'] // 10) * 10 # redondear a la década más cercana
temperature_data.head(11)
```

`matplotlib` tiene dificultades para crear diagramas de caja gestionando el agrupamiento y la representación de los datos de manera interna, por lo que necesitamos agruparlos nosotros mismos:

```python
# agrupar las temperaturas por década
grouped_temperatures = temperature_data.groupby('Decade')['Temperature'].apply(list)
grouped_temperatures
```

Ahora podemos usar la función `plt.boxplot` de `matplotlib` para crear nuestro gráfico:

```python
plt.figure(figsize=(10, 5))
plt.boxplot(grouped_temperatures, labels=grouped_temperatures.index) # el indice es la década
plt.title('Distribución de la temperatura media anual por década')
plt.xlabel('Década')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

¡Este gráfico es mucho más representativo de la variación de la temperatura a lo largo del tiempo, así como de cómo cambia la distribución de cada década!

Si quisiéramos evitar agrupar manualmente las temperaturas, `seaborn` es capaz de agruparlas internamente y crear diagramas de caja utilizando la función `sns.boxplot`. Puedes encontrar más información sobre ello [aquí](https://seaborn.pydata.org/generated/seaborn.boxplot.html). Vamos a crear el mismo gráfico que hicimos anteriormente:

```python
plt.figure(figsize=(12, 5))
sns.set_style("darkgrid")
sns.boxplot(data=temperature_data, x='Decade', y='Temperature')
plt.title('Distribución de la temperatura media anual por década')
plt.xlabel('Década')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=45)
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/temperaturas_caja.png', dpi=300)
plt.show()
```

## **Gráfico de Violín**

Similar al `diagrama de caja`, el `gráfico de violín` muestra la distribución de datos numéricos utilizando **curvas de densidad**. Es una herramienta útil para visualizar y comparar distribuciones entre grupos, e identificar mínimos, máximos y colas en los datos.

Vamos a crear `gráficos de violín` de los datos de temperatura que graficamos anteriormente.

En `matplotlib`, utilizamos `plt.violinplot`, donde puedes especificar si queremos mostrar la `media` o `mediana` de cada distribución. Los parámetros de entrada son diferentes de los de `plt.boxplot`, y no podemos graficar las etiquetas directamente en la función. Puedes encontrar más información sobre esta función y sus argumentos [aquí](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html).

```python
sns.reset_orig() # resetear el estilo de seaborn
plt.figure(figsize=(12, 5))
plt.violinplot(grouped_temperatures, showmedians=True, showmeans=False) # mostrar la mediana pero no la media
plt.title('Distribución de la temperatura media anual por década')
plt.xlabel('Década')
plt.ylabel('Temperatura (°C)')
plt.xticks(ticks=range(1, len(grouped_temperatures.index) + 1), labels=grouped_temperatures.index, # etiquetas de las décadas
            rotation=45, ha='right')
plt.show()
```

En `seaborn`, la función que utilizamos es `sns.violinplot`, que puede tener considerablemente más argumentos que en `matplotlib`. Puedes encontrar toda la información sobre esta función [aquí](https://seaborn.pydata.org/generated/seaborn.violinplot.html).

Una ventaja de esta función es que, al igual que los diagramas de caja, es capaz de agruparlos internamente, por lo que podemos darle directamente el `temperature_data` original en lugar de los `grouped_temperatures`.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
sns.violinplot(data=temperature_data, x='Decade', y='Temperature')
plt.title('Distribución de la temperatura media anual por década')
plt.xlabel('Década')
plt.ylabel('Temperatura (°C)')
plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/temperatura_violin.png', dpi=300)
plt.show()
```

### **¿Cuándo Usar un Diagrama de Violín en Lugar de un Diagrama de Caja?**

Un `diagrama de violín` es útil cuando queremos mostrar la **forma de la distribución** junto con los detalles que ofrece un diagrama de caja. Es particularmente valioso para conjuntos de datos más grandes, ya que proporciona más información sobre la distribución de los datos.

Por otro lado, un `diagrama de caja` es beneficioso para resaltar la **tendencia central** y la **dispersión de los datos**. También es útil para identificar cualquier `valor atípico` presente en el conjunto de datos.


## **Mapas de Calor**

Estos son una representación **bidimensional** de datos en la que varios valores se representan mediante colores. Usar `mapas de calor` es una buena manera de representar datos de una manera más visual. Todos los `mapas de calor` comparten una cosa en común: utilizan **diferentes colores para representar diferentes valores** y para comunicar las relaciones que pueden existir entre las variables.

En `matplotlib`, la función que puede hacer esto es `plt.imshow`, que no se usa exclusivamente para crear mapas de calor. `plt.imshow` puede mostrar imágenes importadas, interpolar imágenes, crear gráficos de contorno, etc. Puedes encontrar más información sobre esta función [aquí](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html), así como ejemplos adicionales sobre cómo [graficar imágenes](https://matplotlib.org/stable/gallery/images_contours_and_fields/image_demo.html#sphx-glr-gallery-images-contours-and-fields-image-demo-py).

Así es como usaríamos `plt.imshow`:

```python
data = np.random.random((12, 12)) # generar datos aleatorios de 12x12

plt.figure(figsize=(6, 6))
plt.imshow(data, cmap='viridis', interpolation='nearest') # grafiar los datos como una imagen e interpolar los valores
plt.grid(False)  # desactivar la cuadrícula
plt.colorbar() # mostrar la barra de colores
plt.title("Mapa de calor aleatorio")
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.show()
```

Dado que `plt.imshow` es una función tan versátil, el código para crear un mapa de calor específico anotado puede volverse bastante largo. Aquí hay algunos [ejemplos](https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html#sphx-glr-gallery-images-contours-and-fields-image-annotated-heatmap-py) sobre cómo hacerlo.

Afortunadamente, `seaborn` tiene la función `sns.heatmap`, que puede graficar datos rectangulares como una matriz codificada por colores directamente. Como ejemplo, importemos un dataframe del [repositorio de datos de seaborn](https://github.com/mwaskom/seaborn-data/tree/master) con datos sobre las características de los diamantes y creemos un mapa de calor que nos informe sobre el `precio` de un diamante según su `corte`, `color` y `claridad`.

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

Primero separaremos por `corte` y `claridad`. Para esto, utilizaremos el método `pivot_table` de `pandas`, donde las filas estarán indexadas por `corte`, las columnas estarán definidas por `claridad` y los valores de color del mapa de calor serán el `precio` del diamante. Puedes encontrar más información sobre el `pivot_table` en la documentación de `pandas` [aquí](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html).


```python
diamond_cut_clarity = diamonds.pivot_table(index="cut", columns="clarity", values="price")
diamond_cut_clarity
```

¡Ahora podemos avanzar y trazar nuestro mapa de calor usando `sns.heatmap`!

Podemos agregar el valor del precio a la matriz añadiendo el argumento `annot=True`. Además, podemos elegir el `colormap` de nuestro mapa de calor con `cmap`. Puedes encontrar más información sobre esta función [aquí](https://seaborn.pydata.org/generated/seaborn.heatmap.html).

```python
plt.figure(figsize=(12, 6))
sns.heatmap(diamond_cut_clarity, annot=True, cmap='cubehelix', fmt='.0f')
# fmt='.0f' es para formatear la anotación a 0 decimales
plt.title('Precio promedio de diamantes por claridad y corte')
plt.xlabel('Claridad')
plt.ylabel('Corte')
plt.show()
```

Vamos a crear un subplot donde podemos crear mapas de calor que muestren el `precio` de un diamante en función del `corte`, `color` y `claridad`:

```python
diamond_cut_color = diamonds.pivot_table(index="cut", columns="color", values="price")
diamond_clarity_color = diamonds.pivot_table(index="clarity", columns="color", values="price")

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5)) # 1 fila, 3 columnas

# corte y claridad
sns.heatmap(diamond_cut_clarity, annot=True, fmt='.0f', ax=ax1)
ax1.set_title('Precio de los diamantes por corte y claridad')
ax1.set_xlabel('Claridad')
ax1.set_ylabel('Corte')

# cut and color
sns.heatmap(diamond_cut_color, annot=True, fmt='.0f', ax=ax2)
ax2.set_title('Precio de los diamantes por corte y color')
ax2.set_xlabel('Color')
ax2.set_ylabel('Corte')

# clarity and color
sns.heatmap(diamond_clarity_color, annot=True, fmt='.0f', ax=ax3)
ax3.set_title('Precio de los diamantes por claridad y color')
ax3.set_xlabel('Color')
ax3.set_ylabel('Claridad')

plt.savefig('/content/drive/MyDrive/Coding Notebooks /Lesson4 - Plotting/plots/diamantes_mapacalor.png', dpi=300)
plt.show()
```
