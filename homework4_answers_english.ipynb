```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# load penguins dataset
penguins = sns.load_dataset("penguins")
sns.set_style("whitegrid")
```

### **Question 1: Count Plot**

- Create a [count plot](https://seaborn.pydata.org/generated/seaborn.countplot.html) using `seaborn` to display the distribution of the number of penguins across different islands (`island` column).

- **Objective:** Research a new plotting function and visualize how to represent categorical data counts.

*Questions to Answer:*
1. Which island has the largest population of penguins?
2. Which penguin species is present in all three islands?

```python
# Count plot for species across islands
sns.countplot(data=penguins, x='island', hue='species')
plt.title('Penguin Species Distribution by Island')
plt.show()
```

*Answers:* 
1. Biscoe Island has the largest population of penguins 
    - Torgersen = ~50
    - Dream = ~120
    - Biscoe = ~160
2. Adelie is present in all three islands

### **Question 2: Heatmap of Correlations**

- Compute the correlation matrix of the `penguins` dataset for the numerical columns (e.g., bill length, bill depth, flipper length, body mass). For this:
    - Make sure the penguins data frame doesn't have any missing values
    - Find the correlation matrix using the `.corr()` method from `pandas`
- Visualize the correlations using a **heatmap** with `seaborn`. Annotate the heatmap to display the correlation values on the grid.

- **Objective:** Research correlation matrix in pandas and learn to use heatmaps to understand variable relationships.

*Questions to Answer:*
1. Which variables have the strongest positive correlation?
2. Are there any variables that show little to no correlation?

```python
# drop missing values
penguins_clean = penguins.dropna()

# correlation matrix
corr_matrix = penguins_clean[['bill_length_mm', 'bill_depth_mm', 
                            'flipper_length_mm', 'body_mass_g']].corr()

# heatmap of correlations
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Heatmap of Penguin Measurements')
plt.show()
```

*Answers:* 
1. Flipper length and body mass show the strongest positive correlation (around 0.87)
2. Bill depth and flipper length show very little correlation (close to 0)

### **Question 3: Pie Chart – Species Distribution**

- Using `matplotlib`, create a [pie chart](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html) showing the proportion of each species in the dataset.

- **Objective:** Visualize categorical data using a pie chart.

*Questions to Answer:*
1. Which species makes up the largest proportion of the population?

```python
# Counting species
species_counts = penguins['species'].value_counts()

# Pie chart for species distribution
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Species Distribution of Penguins')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

*Answers:* 
1. Adélie penguins make up the largest proportion of the penguin population in the dataset, followed by Gentoo and then Chinstrap penguins.
