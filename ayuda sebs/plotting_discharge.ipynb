## Plotting Erlenbach Discharge 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# define the file path
file_path = 'C:/Users/nicol/Documents/GitHub/colab_test/ayuda sebs/qu-20240701-20240818.lst'
# load the data
df = pd.read_csv(file_path, sep='\s+', skiprows=2, header=None)
# combine the date and time columns (if split) into a single datetime string
df['DATUM'] = df[0] + ' ' + df[1]
# rename the discharge column
df.rename(columns={2: 'QU'}, inplace=True)
# convert the 'DATUM' column to datetime format
df['DATUM'] = pd.to_datetime(df['DATUM'], format='%d.%m.%Y %H:%M')
# drop the extra columns
df = df[['DATUM', 'QU']]

print(df)
```

```python
# Plotting the discharge over time
plt.figure(figsize=(10, 6))
plt.plot(df['DATUM'], df['QU'], linestyle='-', color='b')
plt.title('Erlenbach Discharge')
plt.xlabel('Time')
plt.ylabel('Discharge')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('Erlenbach_Discharge.png', dpi=500)

# Display the plot
plt.show()
```

Only plotting august

```python
# Filter the DataFrame to include only rows where the month is August
df_august = df[df['DATUM'].dt.month == 8]

# Plotting the discharge over time for August
plt.figure(figsize=(10, 6))
plt.plot(df_august['DATUM'], df_august['QU'], linestyle='-', color='b')
plt.title('Discharge in August')
plt.xlabel('Time')
plt.ylabel('Discharge (mÂ³/s)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('August_Discharge.png', dpi=500)
# Display the plot
plt.show()
```

```python
# Define the date range for filtering
start_date = '2024-08-18'
end_date = '2024-08-19'

# Filter the DataFrame to include only rows within the specified date range
df_range = df[(df['DATUM'] >= start_date) & (df['DATUM'] <= end_date)]

# Plotting the discharge over time for the specified date range
plt.figure(figsize=(10, 6))
plt.plot(df_range['DATUM'], df_range['QU'], marker='o', linestyle='-', color='b')
plt.title(f'Discharge from {start_date} to {end_date}')
plt.xlabel('Time')
plt.ylabel('Discharge')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('high_flow_event.png', dpi=500)

# Display the plot
plt.show()
```
